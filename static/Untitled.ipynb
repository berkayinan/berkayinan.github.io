```python
import librosa
import matplotlib.pyplot as plt
import json
%matplotlib inline
```

```python
wav_form_mixed,sr = librosa.load('./john_vs_stephen_h264_full.mp4',sr=16000)
wav_form_clean,sr = librosa.load('./john_olvier_fb.mp4',sr=16000)

wav_form_out,sr = librosa.load('./john_vs_stephen_h264_full.mp4',sr=16000)
wav_form_inter,sr = librosa.load('./stephen_colbert_16.mp4',sr=16000)
```

```python
wav_form_clean.shape[0]//100
```

```python
def to_str_list(npa):
    return [round(x,3) for x in npa.tolist()][::200]
```

```python
json_dict = {'on-screen':to_str_list(wav_form_clean),'off-screen':to_str_list(wav_form_inter)}
with open('waveform_colbert_vs_stephen.json','w') as json_fp:
    json.dump(json_dict,json_fp)
```

```python
# plt.plot(wav_form_mixed)
plt.plot(wav_form_clean,alpha=0.5)

plt.plot(wav_form_inter,alpha=0.5)
plt.grid()
plt.ylim([-0.5,0.5])

# plt.plot(wav_form_inter)
```
